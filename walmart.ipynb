```python
# Create a function that changes the datatypes of columns.
def change_col_type(df):
    assign_types = {'Store':'category', 'Holiday_Flag':'category',
                    'Fuel_Price':'float16', 'CPI':'float16',
                    'Temperature':'float16', 'Unemployment':'float16'}
    return (df
            .astype(assign_types)
            .assign(Date=pd.to_datetime(df.Date, format='%d-%m-%Y'))
            )

df = change_col_type(data)
```

```python
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Check how variables are correlated.
fig, ax = plt.subplots(figsize=(10, 4), dpi=300)
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', ax=ax)
ax.set_title('Correlation Plot', fontweight='bold');
```
